Criamos um novo dataframe chamado df_consulta3 que é uma cópia do dataframe df_consulta, mas sem as colunas 'mes' e 'ano'.

 A função drop é usada para remover colunas ou linhas do dataframe.

 O parâmetro columns especifica quais colunas devem ser removidas

```python
df_consulta3=df_consulta.drop(columns=['mes','ano'])
```

Calculamos a média de todas as colunas numéricas de um dataframe através da variável mean

E armazena o resultado na variável media

```python
media=df_consulta3.mean()
display(media)
```

Calcula a mediana de todas as colunas numéricas do dataframe df_consulta3

```python
mediana=df_consulta3.median()
display(media)
```

Calcula os quartis Q1 (25%), Q2 (50%) e Q3 (75%) do dataframe df_consulta3 e exibe os resultados.

```python
q1=df_consulta3.quantile(0.25)
q2=df_consulta3.quantile(0.50)
q3=df_consulta3.quantile(0.75)

print(f"  Q1 (25%): {q1}")
print(f"  Q3 (50%): {q2}")
print(f"  Q3 (75%): {q3}")  
```

Calcula o intervalo interquartil (IQR) subtraindo o primeiro quartil (Q1) do terceiro quartil (Q3) e exibe o resultado

```python
iqr=q3-q1
print(iqr)
```

Calcula os limites superior e inferior usando o intervalo interquartil (IQR) e exibe os resultados.

Calcula o limite superior adicionando 1,5 vezes o intervalo interquartil (IQR) ao terceiro quartil (Q3).

Calcula o limite inferior subtraindo 1,5 vezes o intervalo interquartil (IQR) do primeiro quartil (Q1)

Este valor é usado para identificar possíveis valores atípicos abaixo do intervalo interquartil.

```python
limite_superior=q3+(1.5*iqr)
limite_inferior=q1-(1.5*iqr)
print(f'Limite Superior: {limite_superior}')
print(f'Limite Inferior: {limite_inferior}')
```

Calcula a distância relativa entre a média e a mediana, dividindo a diferença pela mediana,

```python
distancia=(media-mediana)/mediana
print(distancia)
```

Calcula a variância de todas as colunas numéricas do dataframe df_consulta3

```python
variancia=df_consulta3.var()
display(variancia)
```

Calcula o desvio padrão de todas as colunas numéricas do dataframe

```python
desviopadrao=df_consulta3.std()
display(desviopadrao)
```

Calcula uma segunda distancia,agora entre o desvio Padrão dividido pela média ao quadrado 

```python
distancia2=(df_consulta3.std())/((df_consulta3.mean())**2)
display(distancia2)
```

Calcula o valor minimo de cada coluna no do DataFrame.

```python
valor_minimo=df_consulta3.min()
print(f'Menor Valor:{valor_minimo}')
```

Calcula o valor Máximo de cada coluna no do DataFrame.

```python
valor_maximo=df_consulta3.max()
print(f'Maior Valor:{valor_maximo}')
```
